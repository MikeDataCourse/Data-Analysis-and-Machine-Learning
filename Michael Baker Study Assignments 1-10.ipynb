```python
# Find the amount of tips received each day
total_tips_by_day = tips.groupby('day')['tip'].sum()
print(total_tips_by_day)
```

```python
# What is the average total bill for groups larger than 4 people?
average_total_bill = tips[tips['size'] >4]['total_bill'].mean()
print(average_total_bill)
```

```python
# Group by time(Lunch or dinner) and calculate the average tip percentage
tips['tip_percentage'] = tips['tip'] / tips['total_bill'] * 100 
average_tip_percentage = tips.groupby('time')['tip_percentage'].mean()
print(average_tip_percentage)
```

```python
plt.hist(tips['total_bill'], bins=20, color='skyblue', edgecolor='blue')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.title('Distribution of Total Bill in Tips Dataset')
plt.show()
```

```python
# Plotting a bar plot for average tips by day
average_tips_by_day = tips.groupby('day') ['tip'].mean().reset_index()
plt.figure(figsize=(8, 5))
sns.barplot(data=average_tips_by_day, x='day', y='tip', color='skyblue')
plt.xlabel('Day')
plt.ylabel('Average Tip')
plt.title('Average Tips by Day')
plt.show()
```

```python
# Load the Titanic Dataset
titanic =sns.load_dataset('titanic')
print(titanic.head())
```

```python
# Check for missing values
# Check for missing values in each column
missing_values = titanic.isnull().sum()

# Display the missing values count for each column
print(missing_values)
```

```python
#Fill missing values in the age column with median age.
# Fill missing values in the 'age' column with the median value of 'age'
titanic['age'] = titanic['age'].fillna(titanic['age'].median())

# Check if there are any remaining missing values in the 'age' column
print(titanic['age'].isnull().sum())
```

```python

# Drop rows where the 'embarked' column has missing values
titanic_cleaned = titanic.dropna(subset=['embarked'])

# Check the number of rows before and after dropping
print(f"Rows before dropping: {len(titanic)}")
print(f"Rows after dropping: {len(titanic_cleaned)}")
```

```python
# Define a function to categorize passengers by age
def categorize_age(age):
    if age < 18:
        return 'Child'
    elif 18 <= age < 60:
        return 'Adult'
    else:
        return 'Senior'

# Apply the function to create a new 'age_group' column
titanic['age_group'] = titanic['age'].apply(categorize_age)

# Display the first few rows to check the new column
print(titanic[['age', 'age_group']].head())
```

```python
#Assignment 9 
def calculate_tip_percentage(total_bill, tip):
    if total_bill == 0:
        return 0  # To handle division by zero
    tip_percentage = (tip / total_bill) * 100
    return tip_percentage
```
